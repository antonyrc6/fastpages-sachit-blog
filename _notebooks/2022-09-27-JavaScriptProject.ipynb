# JavaScript Fibonacci Sequence Project

I created this project to show what I learned about JavaScript this week.

```javascript
function myFibonacci(sequence, amount) {
  for (let i = 0; i < amount; i++) {
    let lastItem = sequence[sequence.length - 1];
    let secondToLastItem = sequence[sequence.length - 2];
    sequence.push(lastItem + secondToLastItem);
  }
  return sequence;
}
console.log(myFibonacci([1, 1, 2], 97));
```
